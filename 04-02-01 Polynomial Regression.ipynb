## 01 : Data Mining

```python
# Generating Data
data = [[1,45000],[2,50000],[3,60000],[4,80000],[5,110000],[6,150000],[7,200000],[8,300000],[9,500000],[10,1000000]]
data = pd.DataFrame(data)
data.columns = ["yr_exp","salary"]
```
